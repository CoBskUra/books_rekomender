```python
import pandas as pd
import random
import math
```

```python
books = pd.read_csv('orginal_df_books.csv', on_bad_lines='skip')
books.head()
```

# 1 Dosotosowanie dazy danych do naszych potrzeb

## 1.1 Usuwanie zbędnych kolumn

```python
colums_to_delete = {"isbn", "isbn13", "text_reviews_count"}
for column_name in colums_to_delete:
    books.drop(column_name, inplace=True, axis=1)
```

```python
books
```

## 1.2 Dodawanie kolumn

```python
def Make_List_From_txt(title: str):
    genres = open('additional_data_base_info/' + title + '.txt','r')
    genres_list =[]
    for line in genres:
        genres_list.append(line.strip())
    return genres_list
    
def Create_Extra_Column(column_name: str, min: int, max: int, content_list: list, seed: int = 0):
    random.seed(seed)
    rows_number = books.shape[0]+1
    new_column = []
    for id in range(1,rows_number):
        element_number = random.randint(min, max)
        row_content = random.choices(content_list,k = element_number)
        if min == 1 and max == 1:
            new_column.append(row_content[0])
        else:
            new_column.append(row_content)
    books[column_name] = new_column
```

```python
genres = Make_List_From_txt("genres")
target_groups = Make_List_From_txt("target_groups")
tags = Make_List_From_txt("tags")
countrys = Make_List_From_txt("country_of_origin")

Create_Extra_Column("Genres", 1, 3, genres)
Create_Extra_Column("target_groups", 1, 1, target_groups)
Create_Extra_Column("tags", 1, 7, tags)
Create_Extra_Column("country_of_origin", 1, 1, countrys)


# liczba wyporzyczeń w miesiącu
random.seed(0)
books['month_rentals'] = (books["ratings_count"] / random.randint(10,15)).astype(int)

books.head()
```

## 1.3 zmiana kolejności kolumn oraz nazw kolumn

```python
books = books[["bookID", "title", "authors", "publisher", "Genres", "target_groups", "tags", "  num_pages", "language_code", "country_of_origin", "publication_date", "average_rating", "ratings_count", "month_rentals"]]

# zmiana nazwy
books.rename(columns = {'ratings_count':'ratings_count', '  num_pages':'num_pages'}, inplace = True)



books.head()
```

## 1.4 Zmiana zawartości kolumn 

```python
books["bookID"] = range(0, books.count()[0])
```

# Zapis danych

```python
books.to_csv("books.csv", index = False)
```
